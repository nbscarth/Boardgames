```python
# Imports
from splinter import Browser
from bs4 import BeautifulSoup as soup
import pandas as pd
```

```python
# Open browser
browser = Browser("chrome")
```

```python
# Parse html from webpage
url = "https://boardgamegeek.com/browse/boardgame"
browser.visit(url)

html = browser.html

page = soup(html, 'html.parser')
```

```python
# Find all table rows
rows = page.find_all('tr', id='row_')
rows
```

```python
# Test for loop to extract row information
for row in rows:
    rank = row.find('a')['name']
    name = row.find('a', class_='primary').text
    link = row.find('a', class_='primary')['href']
    print(f"Rank: {rank}, Title: {name}, Link: {link}")
```

```python
# Create list of dictionaries for dataframe
table_list = []
for row in rows:
    rank = row.find('a')['name']
    name = row.find('a', class_='primary').text
    link = row.find('a', class_='primary')['href']
    
    row_dict = {
        'rank': rank,
        'title': name,
        'link': link
                }
    table_list.append(row_dict)
```

```python
# Create dataframe
df = pd.DataFrame(table_list)
df
```

```python
# Convert link to game ID#
df['game id'] = df['link'].str.extract(r"/(\d+)/")
new_df = df[['rank', 'title', 'game id']]
new_df.head()
```

```python
new_df.to_csv('initial_scrape_dataframe', index=False)
```
